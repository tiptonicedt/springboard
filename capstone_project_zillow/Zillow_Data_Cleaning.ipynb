```python
#importing Zillow timeseries housing data 
homes2 = pd.DataFrame.from_csv('data/Zip_Zhvi_AllHomes.csv', index_col=None)
homes2.head()
```

```python
#importing household income data by zip code
hhi = pd.DataFrame.from_csv('data/Income_by_zip.csv', index_col=None)
hhi.columns = ('RegionName', 'MedianIncome', 'MeanIncome', 'Pop')
hhi.head()
```

```python
df = pd.merge(homes2, hhi, how='left', left_on='RegionName', right_on='RegionName')
df.head(10)
```

```python
#joining the data sets
df = homes2.join(hhi, on='RegionName', how='left', lsuffix= 'str', rsuffix= 'str', sort=False)
df.head(50)
```

```python
#attempting to convert MeanIncome to numeric values
import math

def convert_income2(x):
    if not isinstance(x, str):
        return x
    elif x == '.':
        return None
    elif math.isnan(float(x.replace(',', ''))):
        return None
    else:
        return int(x.replace(',', ''))

df['MeanIncome'] = df['MeanIncome'].apply(convert_income2)
```

```python
coastal_states = 'HI,RI,CA,WA,OR,TX,LA,FL,NC,SC,GA,VA,MD,DE,NJ,NY,MS,CT,MA,ME,AK'.split(',')
df['Coastal'] = np.where(df[df.State.isin(coastal_states)], 'True', 'False')
df.head()
```

```python
pacific_states = 'HI,CA,WA,OR,AK'.split(',')
df_pacific = df[df.State.isin(pacific_states)]
atlantic_states = 'RI,FL,NC,SC,GA,VA,MD,DE,NJ,NY,CT,MA,ME'.split(',')
df_atlantic = df[df.State.isin(atlantic_states)]
```
